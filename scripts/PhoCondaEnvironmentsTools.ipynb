```python
"""
Tools for managing Conda (Anaconda) and Mamba environments
Include:
    - Accessing revisions/history
"""
import os
import sys
from pathlib import Path
```

```python
envs_path = Path(r'C:\Users\pho\miniconda3\envs')
envs_path
```

```python
def get_historyfile_path(env_path):
    return env_path.joinpath('conda-meta','history')
    
env_paths = [f for f in envs_path.iterdir() if f.is_dir()]
env_paths

env_historfiles_paths = [get_historyfile_path(f) for f in env_paths]
env_historfiles_paths
```

```python
output_history_files = dict()
for env_path, historyfile_path in zip(env_paths, env_historfiles_paths):
    env_name = env_path.name
    try:
        historyfile_contents = historyfile_path.read_text()
    except FileNotFoundError as e:
        print(f'{e}')
        historyfile_contents = None
    output_history_files[env_name] = historyfile_contents
output_history_files
```

```python
def fast_scandir(dirname):
    subfolders= [f.path for f in os.scandir(dirname) if f.is_dir()]
    for dirname in list(subfolders):
        subfolders.extend(fast_scandir(dirname))
    return subfolders
```

```python
fast_scandir(envs_path)
```
